## Yay Neural Networks

neuron with 3 inputs

```python
inputs= [2.5, 1.8, 3.9]
weights =[
    [1.3, 1.7, 2.9],
    [1.7, 4.2, 1.1],
    [6.2, 1.3, 0.7]
]
biases=[1.2,1.6,1.9]

outputs=[]
for neuron_weights, neuron_biases in zip(weights, biases):
    neuron_output=0
    for weight, input in zip(neuron_weights,inputs):
        neuron_output+=weight*input
    neuron_output+=neuron_biases
    outputs.append(neuron_output)
outputs
```

## using the np

```python
import numpy as np
```

```python
inputs= [2.5, 1.8, 3.9]
weights =[
    [1.3, 1.7, 2.9],
    [1.7, 4.2, 1.1],
    [6.2, 1.3, 0.7]
]
biases=[1.2,1.6,1.9]#iterates through the weights list
output = np.dot(inputs,weights)+biases
print(output)
```

## Batch processing

```python
inputs= [[2.5, 1.8, 3.9,2],
        [2.0,5.0,-1.0,4],
        [-1.5,2.7,2.2,5]] 
weights =[
    [1.3, 1.7, 2.9, 1.9],
    [1.7, 4.2, 1.1, 1.9],
    [6.2, 1.3, 0.7, 1.9]
]
biases=[1.2,1.6,1.9]
output = np.dot(inputs,np.array(weights).T)+biases
#why transform
#no of inputs in each batch should be equal to number of lists of weights
print(output)
```

## layering

```python
inputs= [[2.5, 1.8, 3.9,2],
        [2.0,5.0,-1.0,4],
        [-1.5,2.7,2.2,5]] 
weights_1 =[
    [1.3, 1.7, 2.9, 1.9],
    [1.7, 4.2, 1.1, 1.9],
    [6.2, 1.3, 0.7, 1.9]
]
biases_1=[1.2,1.6,1.9]
layer_1_output = np.dot(inputs,np.array(weights_1).T)+biases_1

#layer2
weights_2 =[
    [1.7, 1.7, 2.9],
    [1.3, 6.1, 8.7],
    [1.8, 1.3, -1.2]
]
biases_2=[6.3,1.8,2.1]
layer_2_output = np.dot(layer_1_output,np.array(weights_2).T)+biases_2
print(layer_2_output)
```
